```python
#our aim is to predict net hourly electrical output
import pandas as pd
import numpy as np
import tensorflow as tf
```

# **DATA PREPROCESSING**

```python
data = pd.read_excel("Folds5x2_pp.xlsx")
data.head()
```

```python
#"iloc" is used to select indexes of thw column we want to select
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values
```

```python
#splitting dataset into test and train
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.20, random_state = 0)
```

# **BUILDING ANN**

```python
import tensorflow as tf
#tensorflow is a deep learning library which allows to build complex neural network
#and train them with tools like optimizer or lost function or dropout tools


#ANN has 2 types
#1 sequence of layers
#2 computational graph
```

```python
#initialize ANN as sequence of layers
ann = tf.keras.models.Sequential()
```

```python
#adding input layer and first hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
```

```python
#adding second hidden layer
ann.add(tf.keras.layers.Dense(units = 6, activation = "relu"))
```

```python
#adding output layer
ann.add(tf.keras.layers.Dense(units = 1, activation = None))
```

# **TRAINING ANN**

```python
#compiling ANN with optimizer and loss function
#connecting our ANN and optimizer and loss function
ann.compile(optimizer = "adam", loss = "mean_squared_error")
```

```python
#training ANN in training set using fit method
ann.fit(x_train, y_train, batch_size = 32, epochs = 100)
```

```python
#predicting results of test set
y_pred = ann.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)), 1))  # 1 for vertical concatnation and 0 for horizontal
```
